```python
import pandas as pd
import requests
from bs4 import BeautifulSoup


def get_wiki_content_without_titles(url):
    """Fetches the content of a Wikipedia article without titles using web scraping.

    Args:
        url: The URL of the Wikipedia article to scrape.

    Returns:
        A string containing the extracted content of the article (excluding titles).
    """

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all paragraphs within the main content
    main_content = soup.find(id="bodyContent")
    paragraphs = main_content.find_all("p")

    # Loop through paragraphs and extract text, excluding headings
    content = ""
    for paragraph in paragraphs:
        if not paragraph.find("h"):  # Check if paragraph contains any heading tag (h1, h2, etc.)
            content += paragraph.get_text() + "\n"  # Add newline between paragraphs

    return content
```
